```python
import pandas as pd
import numpy as np 
import seaborn as sns
import geopandas as gpd

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.colors import ListedColormap
from matplotlib.colors import LinearSegmentedColormap
from matplotlib.colorbar import ColorbarBase 
from matplotlib import colormaps as cm
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize

import networkx as nx
import pyballmapper as pbm

from sklearn.preprocessing import MinMaxScaler
```

```python
# Replace with your actual path if different
file_path = "/Users/alexander/Documents/MSc Data Science/erp-uom/data/dataTurnoutFinal.csv"

# Load the CSV file
df = pd.read_csv(file_path)
```

```python
# Columns you want to exclude from multiplication
exclude_cols = [
    'religious_fractionalization', 'ethnic_fractionalization',
    'total_spending', 'enp_votes', 'electoral_density'
]

# 1) Full list of target columns from col 7 to 51 (0-based, so col 7 is index 7, col 51 is index 51)
all_cols = df.columns[7:52]  # .columns[7:52] gets columns 7â€“51 inclusive

# Filter to get columns to multiply: all minus excluded ones
cols_to_multiply = [col for col in all_cols if col not in exclude_cols]

# Multiply these columns by 100
df[cols_to_multiply] = df[cols_to_multiply] * 100.0

# Variables to scale from 0 to 100
vars_to_scale = [
    'religious_fractionalization', 'ethnic_fractionalization',
    'total_spending', 'enp_votes', 'electoral_density'
]

# Function to apply MinMaxScaler and return a DataFrame
def minmax_scale_0_100(group):
    scaler = MinMaxScaler()
    group[vars_to_scale] = scaler.fit_transform(group[vars_to_scale]) * 100.0
    return group

# Apply per election year
df = df.groupby('election_year', group_keys=False, observed=True).apply(
    minmax_scale_0_100
)
```

## Descriptive statistics for full topology

```python
# 1. Define colour map
# Ball Mapper Topology (2010) â€” eps = 70
hsvp = cm.get_cmap("gist_rainbow")
vmin, vmax = 0.40, 1.00
norm = mcolors.Normalize(vmin=vmin, vmax=vmax)

# 2. Create filtered DataFrame for 2010
df_2010 = df[df['election_year'] == 2010].reset_index(drop=True)
X = df_2010.iloc[:, 7:]
color_df = df_2010[['electoralTurnout']]

# 3. Create BallMapper object
bm2010 = pbm.BallMapper(X=X, eps=70, coloring_df=color_df)

# 4. Assign node colours using gist_rainbow
for node in bm2010.Graph.nodes:
    val = bm2010.Graph.nodes[node]['electoralTurnout']
    if val is not None and not np.isnan(val):
        bm2010.Graph.nodes[node]['color'] = hsvp(norm(val))
    else:
        bm2010.Graph.nodes[node]['color'] = "black"

# 5. Plot the topology
plt.figure(figsize=(10, 7))
bm2010.draw_networkx(
    coloring_variable='electoralTurnout',
    color_palette=hsvp,
    colorbar=True,
    vmin=vmin,
    vmax=vmax
)
plt.show()

# Extract balls and positions
pb2010 = bm2010.points_and_balls() 
df_2010['point'] = df_2010.index
points2010 = df_2010.merge(pb2010, on='point')

ball_summary = points2010.groupby('ball').agg(
    n_constituencies=('point', 'count'),
    turnout_mean=('electoralTurnout', 'mean')
).reset_index()

feature_cols = df_2010.columns[7:-1]  # exclude 'point'
feature_means = points2010.groupby('ball')[feature_cols].mean().reset_index()

ball_stats2010 = pd.merge(ball_summary, feature_means, on='ball')
ball_stats2010['election_year'] = 2010
```

```python
# Step 1: Keep per-year means and SDs (already done)
mean_sd_table = df.groupby('election_year')['electoralTurnout'].agg(['mean', 'std']).reset_index()

# Step 2: Calculate global mean and SD
global_mean = df['electoralTurnout'].mean()
global_sd = df['electoralTurnout'].std()

# Step 3: Local classification function (per year)
def tag_turnout_category(stats_df, year):
    mean_val = mean_sd_table.loc[mean_sd_table['election_year'] == year, 'mean'].values[0]
    std_val = mean_sd_table.loc[mean_sd_table['election_year'] == year, 'std'].values[0]

    def label(row):
        if row['turnout_mean'] < mean_val - std_val:
            return 'low'
        elif row['turnout_mean'] > mean_val + std_val:
            return 'high'
        else:
            return 'normal'

    stats_df['turnout_category'] = stats_df.apply(label, axis=1)
    return stats_df

# Step 4: Global classification function
def tag_turnout_category_global(stats_df):
    def label(row):
        if row['turnout_mean'] < global_mean - global_sd:
            return 'low'
        elif row['turnout_mean'] > global_mean + global_sd:
            return 'high'
        else:
            return 'normal'
    stats_df['turnout_category_global'] = stats_df.apply(label, axis=1)
    return stats_df

# Step 5: Combined classification (either local OR global is low)
def tag_turnout_category_combined(stats_df):
    def combined_label(row):
        if row['turnout_category'] == 'low' or row['turnout_category_global'] == 'low':
            return 'low'
        elif row['turnout_category'] == 'high' or row['turnout_category_global'] == 'high':
            return 'high'
        else:
            return 'normal'
    stats_df['turnout_category_combined'] = stats_df.apply(combined_label, axis=1)
    return stats_df

# Step 6: Apply all three taggers to each DataFrame
def apply_all_labels(stats_df, year):
    stats_df = tag_turnout_category(stats_df, year)
    stats_df = tag_turnout_category_global(stats_df)
    stats_df = tag_turnout_category_combined(stats_df)
    return stats_df

ball_stats2010 = apply_all_labels(ball_stats2010, 2010)
ball_stats2015 = apply_all_labels(ball_stats2015, 2015)
ball_stats2017 = apply_all_labels(ball_stats2017, 2017)
ball_stats2019 = apply_all_labels(ball_stats2019, 2019)

# Step 7: Combine into one
ball_stats_all = pd.concat(
    [ball_stats2010, ball_stats2015, ball_stats2017, ball_stats2019],
    ignore_index=True
)

# Optional export
# ball_stats_all.to_csv("/Users/alexander/Documents/MSc Data Science/erp-uom/data/ball_stats_all_years.csv", index=False)
```

```python
def gini(x):
    x = np.sort(np.array(x))
    n = len(x)
    index = np.arange(1, n + 1)
    return (2 * np.sum(index * x) / np.sum(x) / n) - (n + 1) / n

concentration_table = pd.DataFrame({
    'election_year': [2010, 2015, 2017, 2019],
    'gini_n_constituencies': [
        gini(ball_stats2010['n_constituencies']),
        gini(ball_stats2015['n_constituencies']),
        gini(ball_stats2017['n_constituencies']),
        gini(ball_stats2019['n_constituencies']),
    ]
})

concentration_table
```

```python
# Step 1: Keep only one row per constituency per year
deduped = all_points.drop_duplicates(subset=['election_year', 'constituency_geographic_code'])

# Step 2: Calculate low_any_score and low_only_score
summary = deduped.groupby('constituency_geographic_code').agg(
    low_any_score=('any_low', 'sum'),
    low_only_score=('only_low', 'sum')
).reset_index()

# Step 3: Add name for readability
names = all_points[['constituency_geographic_code', 'constituency_name']].drop_duplicates()
summary = summary.merge(names, on='constituency_geographic_code', how='left')

# Step 4: Classify into participation class
def classify(row):
    if row['low_any_score'] == 0:
        return 'never_low'
    elif row['low_only_score'] == 4:  # all years only in low balls
        return 'persistent_low'
    else:
        return 'intermittent_low'

summary['participation_class'] = summary.apply(classify, axis=1)

# Step 5: Reorder for readability
summary = summary[[
    'constituency_geographic_code', 'constituency_name',
    'low_any_score', 'low_only_score', 'participation_class'
]]
```

```python
# Step 1: Keep only one row per constituency per year
deduped = all_points.drop_duplicates(subset=['election_year', 'constituency_geographic_code'])

# Step 2: Calculate low_any_score and low_only_score
summary = deduped.groupby('constituency_geographic_code').agg(
    low_any_score=('any_low', 'sum'),
    low_only_score=('only_low', 'sum')
).reset_index()

# Step 3: Add name for readability
names = all_points[['constituency_geographic_code', 'constituency_name']].drop_duplicates()
summary = summary.merge(names, on='constituency_geographic_code', how='left')

# Step 4: Classify into participation class
def classify(row):
    if row['low_any_score'] == 0:
        return 'never_low'
    elif row['low_only_score'] == 4:  # all years only in low balls
        return 'persistent_low'
    else:
        return 'intermittent_low'

summary['participation_class'] = summary.apply(classify, axis=1)

# Step 5: Reorder for readability
summary = summary[[
    'constituency_geographic_code', 'constituency_name',
    'low_any_score', 'low_only_score', 'participation_class'
]]
```
